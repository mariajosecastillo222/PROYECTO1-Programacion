```python
import random
from time import sleep
```

```python
MAX_PLAYERS = 4
valor_fijo = 0
pozo = 0
PlayersInfo = []
Players = 0
IsWinner = False
```

```python
DadoNumbers = [1, 2, 3, 4, 5, 6]
```

```python
def GetPlayersName():
    global PlayersInfo
    for i in range(Players):
        print("¬°Hola jugador", str(i+1) + "!, por favor ingresa tu nombre:")
        CurrentName = input("")
        PlayersInfo.append([CurrentName, 10000, 0])
        print("El jugador", CurrentName, "se registr√≥ correctamente\n\n")
```

```python
def Dado():
    print("Tirando los dados....")
    sleep(2.0)
    num = random.choice(DadoNumbers)
    print("|---------- DADOS:", num, "----------|")
    return num
```

```python
def TomarApuesta(jugador, valor_fijo,pozo):
    
    if valor_fijo > pozo:
        Apuesta = int(input(f"""Ingrese el monto a apostar: (Apuesta minima: {pozo})\n"""))
    else:
        Apuesta = int(input(f"""Ingrese el monto a apostar: (Apuesta minima: {valor_fijo})\n"""))
    
    if Apuesta < valor_fijo:
        
        print(f"""¬°Por favor ingrese un monto v√°lido! (Apuesta minima: {valor_fijo})""")
        TomarApuesta(jugador, valor_fijo, pozo)
    else:
        return Apuesta
```

```python
DeletedPlayers = []
```

```python
def DeletePlayers():
    global PlayersInfo
    global DeletedPlayers
    for player in DeletedPlayers:
        del PlayersInfo[player]
    DeletedPlayers = []
```

```python
def VerifyPlayers(pozo):
    global DeletedPlayers
    global PlayersInfo
    global Players
    global IsWinner
    
    if pozo <= 0:
        
          
                
        DeletePlayers()
        print("\n| -------------  TENEMOS UN GANADOR -------------  |\n")
        print("| -------------  ¬°Felicidades", PlayersInfo[1][0] + ", haz sido el ganador de esta partida! ------------- |\n\n")
        
        recaudo = (int(0.05*PlayersInfo[1][1])/1)
        saldo_final = int(PlayersInfo[1][1] - recaudo)
        print("Saldo final:", saldo_final)
        print("El casino recolect√≥ en esta ronda", recaudo)
                
        IsWinner = True
    
```

```python
def Game(valor_fijo, pozo):
    global PlayersInfo
    pozo = valor_fijo * Players
    CurrentPlayer = 0
    while not IsWinner:
        print("Es el turno del jugador", PlayersInfo[CurrentPlayer][0])
        print("Saldo disponible:", PlayersInfo[CurrentPlayer][1])
        print("En el pozo hay: ", pozo)
        
        PlayersInfo[CurrentPlayer][2] = Dado()
        sleep(2.0)
        if PlayersInfo[CurrentPlayer][2] == 1:
            print("¬°Has perdido la apuesta!")
            print("Se descontaron", valor_fijo, "de tu saldo")
            PlayersInfo[CurrentPlayer][1] -= valor_fijo
            
            print("Saldo final:", PlayersInfo[CurrentPlayer][1])
            pozo += valor_fijo
            sleep(2.0)
            
        elif PlayersInfo[CurrentPlayer][2] == 6:
            PlayersInfo[CurrentPlayer][1] += valor_fijo
            pozo -= valor_fijo
            print("¬°Has salvado la apuesta! Saltas de turno")
            print("Se agregaron ",valor_fijo, "de tu saldo")
            
            sleep(2.0)
            
        else:
            Apuesta = TomarApuesta(CurrentPlayer, valor_fijo, pozo)
            print("¬°Vuelve a lanzar los dados para intentar ganar!")
            NewNum = Dado()
            sleep(2.0)
            if NewNum > PlayersInfo[CurrentPlayer][2]:
                print("¬°Has ganado la apuesta!")
                print("Se han sumado", Apuesta, "a tu saldo")
                PlayersInfo[CurrentPlayer][1] += Apuesta
                pozo -= Apuesta
                
                
            else:
                print("¬°Has perdido la apuesta!")
                print("Se descontaron", Apuesta, "de tu saldo")
                PlayersInfo[CurrentPlayer][1] -= Apuesta
                pozo += Apuesta
                print("Saldo final:", PlayersInfo[CurrentPlayer][1])
            
        print("\n\n|---------------------- FIN DEL TURNO ----------------------|\n\n")
        sleep(2.0)
        if CurrentPlayer == Players-1:
            CurrentPlayer = 0 
        else:
            CurrentPlayer += 1
        VerifyPlayers(pozo)
```
